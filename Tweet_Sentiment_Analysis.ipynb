```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from scipy.special import softmax
import os
```

```python
tweet = "@ghosty: fuck you"
#preprocess tweet
tweet_words = []
for word in tweet.split(' '):
    if word.startswith('@') and len(word) > 1:
        word = '@user'
    elif word.startswith('http') or word.startswith('https'):
        word = "http"
    tweet_words.append(word)

tweet_proc = " ".join(tweet_words)
#print(tweet_words)
```

```python
#loading the tokenizer and model transformer
rbt = "cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(rbt)
model = AutoModelForSequenceClassification.from_pretrained("cardiffnlp/twitter-roberta-base-sentiment")
Lbls = ['Negative','Neutral','Positive']
```

```python
#Sentiment analysis
encoded_tweet = tokenizer(tweet_proc, return_tensors = 'pt')

#output = model(encoded_tweet['input_ids'],encoded_tweet['attention_masks'])
output = model(**encoded_tweet)

scores = output [0][0].detach().numpy()
#print (scores)

scores = softmax(scores)


'''for i in range(len(scores)):
    l = Lbls[i]
    s=scores[i]
    print(l,s)'''
```

```python
#Get the sentiment with highest score
max_score_index = scores.argmax()
sentiment = Lbls[max_score_index]

# Print the direct sentiment result
print(f"Sentiment: {sentiment} (Confidence: {scores[max_score_index]:.2%})")
```
